## Crear

```python
## Crear usuario
def crear_usuario():
  nombre = input("Escribe tu nombre: ")
  id = nombre.lower()

  if not buscar_usuario(id) == None:
    print("Ese pingГјino ya existe. Intenta con otro nombre.")
    return

  edad = 0.0
  health = 100.0
  comida_fav = input("ВҝCuГЎl es tu comida favorita?: ").lower()

  usuario = {"user":id, "name" : nombre, "age":edad, "food":comida_fav, "health":health}
  usuarios[id] = usuario

  print(f"рҹҗ§ {nombre}, {int(edad)} aГұos pingГјino. Me encanta {comida_fav}.")
  cambiar_usuario(id)
```

## Consultar

```python
# Obtiene usuario por id y devuelve diccionario
def buscar_usuario(id : str) -> dict:
  return usuarios.get(id)
```

```python
# Visualizar chat
def mostrar_tablero_mensajes():
  for i in tablero_mensajes:
    mensaje = f"""
    ___________________________________
    рҹҗ§ {i["name"]} \t ID: {i["id"]}
    {i["message"]}
    {i["time-stamp"]}
    ___________________________________
    """
    print(mensaje)


mostrar_tablero_mensajes()
```

```python
def listar_usuarios():
  tabla = []
  for i, v in enumerate(usuarios):
    datos = ["рҹҗ§",f"{usuarios[v]["name"]}", f"({int(usuarios[v]["age"])})"]
    datos.append(convertir_salud_a_pescados(usuarios[v]["health"]))
    tabla.append(datos)

  print(tabulate(tabla, tablefmt = "plain"))
```

```python
def convertir_salud_a_pescados(salud : int) -> str:
  texto = "рҹҗҹ "
  pescados = int(salud / 100.0 * 10.0)
  for i in range(pescados):
    texto += "="
  return texto

print(convertir_salud_a_pescados(70))

```

## Actualizar

```python
def cambiar_usuario(id : str):
  global usuario_actual
  id = id.lower()
  usuario = buscar_usuario(id)

  if usuario == None:
    print("No encuentro ese nombre.")
    return

  usuario_actual = usuario["name"]
  print(f"Ahora eres {usuario_actual}")
```

```python
# Actualizar nombre de usuario pidiendo el id
def actualizar_usuario(id : str):
  id = id.lower()
  usuario = buscar_usuario(id)

  if usuario == None:
    print("No encuentro ese nombre.")
    return

  nuevo_nombre = input("ВҝCГіmo te llamarГЎs ahora?: ")
  nuevo_id = nuevo_nombre.lower()
  usuarios[nuevo_id] = usuarios.pop(id)
  usuario["name"] = nuevo_nombre
  cambiar_usuario(nuevo_id.lower())

  for i in tablero_mensajes:
    if i["user_id"] == id.lower():
      i["user_id"] = nuevo_nombre.lower()
      i["name"] = nuevo_nombre
      print("Usuario actualizado")
      return

# print(usuarios)
# actualizar_usuario()
# mostrar_tablero_mensajes()
```

```python
def alimentar_usuario(id) -> str:
  id = id.lower()
  usuario = buscar_usuario(id)

  if usuario == None:
    print("No encuentro ese nombre.")
    return

  comida = input(f"ВҝQuГ© le darГЎs a {usuario["name"]}?: ")

  if usuario["food"] == comida.lower():
    usuario["health"] += 10
    crear_mensaje(f"Me encanta {comida}. ВҝCГіmo lo supiste?", usuario["name"])
  else:
    usuario["health"] += 5
    crear_mensaje(f"Bueno, peor es nada.", usuario["name"])
```

```python
def actualizar_mensaje(id : str):
  for i in tablero_mensajes:
    if i["id"] == id:
      nuevo_mensaje = input("ВҝQuГ© vas a decir ahora?: ")
      i["message"] = nuevo_mensaje + " (editado)"
      mostrar_tablero_mensajes()
      return
```

## Eliminar

```python
# Actualizar nombre de usuario pidiendo el id
def eliminar_usuario(id : str):
  global usuario_actual
  id = id.lower()

  usuario = buscar_usuario(id)
  if usuario == None:
    print("No encuentro ese nombre.")
    return

  print(f"рҹӘҰ {usuario["name"]} ahora duerme con los peces")
  usuarios.pop(id)

  if usuario_actual == id:
    if len(usuarios) == 0:
      usuarios.append({"user":"guest", "name":"Guest"})
    usuario_actual = usuarios[0]["name"]

  for i in tablero_mensajes:
    if i["user_id"] == id:
      # Asignamos un valor aleatorio al id para que el usuario no tenga
      # acceso y para evitar duplicados en el chat
      i["user_id"] = "user" + str(randint(0, 1000))
      i["name"] = "PingГјno desaparecido"
      return

# print(usuarios)
# eliminar_usuario()
# mostrar_tablero_mensajes()
```

```python
def eliminar_mensaje(id : str):
  for i in tablero_mensajes:
    if i["id"] == id:
      tablero_mensajes.remove(i)
      return

  print("No encuentro ese mensaje.")
  return

# eliminar_mensaje()
# mostrar_tablero_mensajes()

```

```python
def actualizar_estados_pinguinos():
  global usuarios
  copia_usuarios = usuarios.copy()

  for i, v in enumerate(copia_usuarios):
    usuarios[v]["age"] += (1 / 10.0) #Cada 10 ciclos incrementa edad en 1
    if usuarios[v]["age"] >= 100.0:
      eliminar_usuario(v)
      continue
    usuarios[v]["health"] -= 5
    if usuarios[v]["health"] <= 0:
      usuarios[v]["health"] = 0
      eliminar_usuario(v)

  usuarios = copia_usuarios
```

```python
def leer_comando(entrada : str) -> []:
   parseo = entrada.split(" ")
   comando = parseo[0]
   parseo_args = parseo[1::]
   args = " ".join(parseo_args)

   instrucciones = [comando, args]
   return instrucciones
```

```python

if __name__ == "__main__":
  print(menu_principal)

  while True:

    entrada = input("\n> ")
    instrucciones = leer_comando(entrada)
    comando = instrucciones[0]
    args = instrucciones[1]

    match comando:
      case "create-penguin":
        crear_usuario()
      case "change-penguin":
        cambiar_usuario(args)
      case "noot":
        crear_mensaje(args, usuario_actual)
      case "edit-noot":
        actualizar_mensaje(args)
      case "edit-name":
        actualizar_usuario(args)
      case "feed":
        alimentar_usuario(args)
      case "del-penguin":
        eliminar_usuario(args)
      case "del-noot":
        eliminar_mensaje(args)
      case "display":
        mostrar_tablero_mensajes()
        listar_usuarios()
      case "help":
        print(ayuda)
      case "noot-noot":
        print("Noot noot! (AdiГіs)")
        break
      case _:
        print("No entiendo ese comando. Escribe help para mostrar lista de comandos")

    actualizar_estados_pinguinos()
```
